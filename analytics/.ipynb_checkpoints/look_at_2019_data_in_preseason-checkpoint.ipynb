```python
import numpy as np
import pandas as pd
from datetime import date
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
players_2019 = pd.read_csv('../vaastav_data/data/2019-20/players_raw.csv')
players_2019.drop(columns=[
    'transfers_in', 
    'transfers_out', 
    'transfers_in_event', 
    'transfers_out_event',
    'cost_change_event',
    'cost_change_event_fall',
    'cost_change_start',
    'cost_change_start_fall',
    'news',
    'news_added',
    'photo',
    'chance_of_playing_this_round',
    'ep_next',
    'ep_this',
    'event_points',
    'form',
    'first_name',
    'second_name'
], inplace=True)
```

```python
players_2019.columns
```

```python
players_2019.describe()
```

```python
players_2019_more_than_100_points = players_2019.loc[players_2019['total_points'] > 100].copy()
players_2019_more_than_100_points['points_per_minute'] = players_2019_more_than_100_points['total_points'] / players_2019_more_than_100_points['minutes']
players_2019_more_than_100_points['points_per_million_unit_price'] = players_2019_more_than_100_points['total_points'] / players_2019_more_than_100_points['now_cost']
players_2019_more_than_100_points.sort_values(by='points_per_million_unit_price', ascending=False).head(10)
```

```python
fig = plt.figure(figsize=(12, 10))
fig.patch.set_facecolor('w')
ax = fig.add_subplot(111)
plt.title('Distribution of Fantasy Points per £Million of of Player Price for Each Player Type', loc='left', y=1.04, color='#444444', weight='bold', fontsize=16)
sns.distplot(players_2019_more_than_100_points.loc[players_2019_more_than_100_points['element_type'] == 1]['points_per_million_unit_price'], label='goalies')
sns.distplot(players_2019_more_than_100_points.loc[players_2019_more_than_100_points['element_type'] == 2]['points_per_million_unit_price'], label='defenders')
sns.distplot(players_2019_more_than_100_points.loc[players_2019_more_than_100_points['element_type'] == 3]['points_per_million_unit_price'], label='midfielders')
sns.distplot(players_2019_more_than_100_points.loc[players_2019_more_than_100_points['element_type'] == 4]['points_per_million_unit_price'], label='forwards', color='#ff0000')
plt.legend(loc=(0, 1.01), ncol=4, frameon=False)
```

```python
pd.plotting.scatter_matrix(players_2019_more_than_100_points[[
    'bps',
    'creativity',
    'element_type',
    'ict_index',
    'threat',
    'selected_by_percent',
    'points_per_million_unit_price'
]].dropna(), alpha=0.8, figsize=(15,15), diagonal='kde')
plt.show()
```
